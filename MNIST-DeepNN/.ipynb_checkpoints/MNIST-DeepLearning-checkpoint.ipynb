```python
import numpy as np
import matplotlib.pyplot as plt
import keras
from keras.datasets import mnist #keras has its own mnist dataset
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.utils.np_utils import to_categorical # one hot encoding
import random
```

```python
np.random.seed(0)
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data() #import 60k datasets(images) and their labels and 10k test sets and their labels
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape[0]) #60k labels
```

```python
assert(X_train.shape[0] == y_train.shape[0]), "The number of images is not equal to the number of labels."
assert(X_test.shape[0] == y_test.shape[0]), "The number of images is not equal to the number of labels."
assert(X_train.shape[1:] == (28,28)), "The dimensions of the images are not 28x28"
assert(X_test.shape[1:] == (28,28)), "The dimensions of the images are not 28x28"
```

```python
num_of_samples = []

cols = 5
num_classes = 10   #0 t0 9

fig, axs = plt.subplots(nrows=num_classes, ncols = cols, figsize=(5, 8)) #figure size of each grid in layout
fig.tight_layout() # minimizing overlapping
for i in range(cols):
    for j in range(num_classes):
        x_selected = X_train[y_train == j]
        axs[j][i].imshow(x_selected[random.randint(0, len(x_selected - 1)), :, :], cmap=plt.get_cmap("gray")) #random images from 0 to 60k imgs and converting them to grayscale
        axs[j][i].axis("off") #remove axis for each img
        if i == 2: # setting a title for each class
            axs[j][i].set_title(str(j))
            num_of_samples.append(len(x_selected))            
```

```python
print(num_of_samples) # shows images in each class
plt.figure(figsize=(12, 4))
plt.bar(range(0, num_classes), num_of_samples)
plt.title("Distribution of the training dataset")
plt.xlabel("Class number")
plt.ylabel("Number of images")
```

```python
y_train = to_categorical(y_train, 10) #converting label data to one hot encoded
y_test = to_categorical(y_test, 10)
```

```python
# Normalization is important
X_train = X_train/255 # 255 beacause maximum pixel intensity is 255 ...it help us to keep the value between 0 and 1
X_test = X_test/255
```

```python
# Flattening of image from 28X28 to 1D array of 784
num_pixels = 784
X_train = X_train.reshape(X_train.shape[0], num_pixels)
X_test = X_test.reshape(X_test.shape[0], num_pixels)
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(10, input_dim=num_pixels, activation='relu'))
    model.add(Dense(30, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(num_classes, activation='softmax')) #output layer #in softmax it goes which has highest probability
    model.compile(Adam(lr=0.01), loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
model = create_model()
print(model.summary())
```

```python
history = model.fit(X_train, y_train, validation_split=0.1, epochs = 10, batch_size = 200, verbose = 1, shuffle = 1)
```

```python
plt.plot(history.history['loss']) #training loss
plt.plot(history.history['val_loss']) #validation loss #applied to allready trained model ...fine tune hyperparams
plt.legend(['loss', 'val_loss'])
plt.title('Loss')
plt.xlabel('epoch')

#validation err should be less than train err otherwise overfitting.......try by changing value of number of epochs
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.legend(['accuracy', 'val_accuracy'])
plt.title('Accuracy')
plt.xlabel('epoch')
```

```python
score = model.evaluate(X_test, y_test, verbose=0) #test images are passed
print(type(score))
print('Test score:', score[0]) #loss
print('Test accuracy:', score[1]*100,'%') #accuracy
```

```python
import requests # to make a get req 
from PIL import Image #python image library
url = 'https://www.researchgate.net/profile/Jose_Sempere/publication/221258631/figure/fig1/AS:305526891139075@1449854695342/Handwritten-digit-2.png'
response = requests.get(url, stream=True)
img = Image.open(response.raw)
plt.imshow(img, cmap=plt.get_cmap('gray'))
```

```python
import cv2

img_array = np.asarray(img) #input data to array conversion
#we save to resize it to 28X28 by using opencv
resized = cv2.resize(img_array, (28, 28))
gray_scale = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
image = cv2.bitwise_not(gray_scale) # to convert background black and digit white
plt.imshow(image, cmap=plt.get_cmap('gray'))
```

```python
#normalize this also
image = image/255
image = image.reshape(1, 784)
```

```python
prediction = model.predict_classes(image)
print("predicted digit:", str(prediction))
```
